```python
spread_col=clean_df['Mid-Career 90th Percentile Salary'].subtract(clean_df['Mid-Career 10th Percentile Salary'])
```

```python
clean_df.insert(1, 'Spread', spread_col)
clean_df.head()
```

```python
low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
```

```python
high_potential=clean_df.sort_values('Mid-Career 90th Percentile Salary',ascending=False)
high_potential[['Undergraduate Major','Mid-Career 90th Percentile Salary']].head()
```

```python
highest_spread=clean_df.sort_values('Spread',ascending=False)
highest_spread[['Undergraduate Major','Spread']].head()
```

```python
highest_median_spread=clean_df.sort_values('Mid-Career Median Salary',ascending=False)
highest_median_spread[['Undergraduate Major','Mid-Career Median Salary']].head()
```
